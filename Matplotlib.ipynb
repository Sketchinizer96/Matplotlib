## Matplotlib example workflow

```python
#0. import matplotlib and get is ready for plotting in jupyter
%matplotlib inline
from matplotlib import pyplot as plt
import random

#1. Prepare data
x1 = np.arange(0,100,10)
y1 = [random.randrange(0, 100, 1) for i in range(10)] #creates a list of 10 random numbers

#2. setup plot
fig, ax = plt.subplots(figsize=(5,5)) #(width, height)

#3. Plot data
ax.plot(x1,y1);

#4. Customize plot
ax.set(title="Simple_plot",
      xlabel="x_axis",ylabel="y_axis");

#5. To save the image use the following command

#fig.savefig("Enter the path here")

fig.savefig("sample_fig.jpg")
```

### Two option for subplots

```python
fig,((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(50),np.random.random(50));
ax3.bar(sales.keys(),sales.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2 : 
fig,ax = plt.subplots(nrows = 2,ncols = 2, figsize=(10,5))

# plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(50),np.random.random(50));
ax[1,0].bar(sales.keys(),sales.values());
ax[1,1].hist(np.random.randn(1000));
```

### Pyplot vs Matplotlib oo method
* When plotting something quickly, we can use pyplot method.
* When plotting advance use OO method

```python
over_50 = heart_disease[heart_disease["age"]> 50]
over_50.head()
```

```python
over_50.plot(kind='scatter',x='age',y='chol',c ='target');
```

```python
# OO method
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age', y='chol', c='target', ax=ax);
ax.set_xlim([45,80]);#set limits on x axis
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax.set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol");
ax.legend(*scatter.legend_elements(), title="target");
# add a horizontal_line
ax.axhline(over_50["chol"].mean(), linestyle='--')
```

```python
over_50.head()
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],y=over_50["trestbps"],c=over_50["target"]);
ax.set(title="Heart Disease and trestbps", xlabel="Age", ylabel="trestbps");
ax.legend(*scatter.legend_elements(),title="target");
ax.axhline(over_50["trestbps"].mean(), linestyle="--");
```

```python
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
fig.suptitle('Heart Disease and Factors', fontsize=16);
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax0.set(title="Heart Disease and Cholestrol Levels", ylabel="Cholestrol");
ax0.legend(*scatter.legend_elements(), title="target");
# add a horizontal_line
ax0.axhline(over_50["chol"].mean(), linestyle='--')
scatter = ax1.scatter(x=over_50["age"],y=over_50["trestbps"],c=over_50["target"]);
ax1.set(title="Heart Disease and trestbps", xlabel="Age", ylabel="trestbps");
ax1.legend(*scatter.legend_elements(),title="target");
ax1.axhline(over_50["trestbps"].mean(), linestyle="--");
fig.suptitle('Heart Disease Analysis', fontsize=16);

```
