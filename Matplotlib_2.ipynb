#### Density and Contours

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

# visualizing a three dimensional function
# contour plot using a function z = f(x,y)

def f(x,y):
    return np.sin(x)** 10+ np.cos(10+y*x)*np.cos(x)

# A contour plot can be created with the plt.contour function. It takes three arguments:
# a grid of x values, a grid of y values, and a grid of z values. The x and y values
# represent positions on the plot, and the z values will be represented by the contour
# levels.

x = np.linspace(0,5,50)
y = np.linspace(0,5,50)
X, Y = np.meshgrid(x,y)
Z = f(X,Y)
# plotting on standard line only contour plot
plt.contour(X,Y,Z,colors='black');

# by default when a single color is used, negative values are represented by
# dashed lines, and positive values by solid lines.
```

```python
plt.contour(X,Y,Z,20,cmap='Accent')
plt.colorbar();
```

```python
#plt.cm.<tab> # use this command to view ALL CMAP OPTIONS
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
            cmap='RdGy')
# plt.imshow() doesnt accept an x and y grid, so you must manually specify the
# extent [xmin, xmax, ymin, ymax] of the image on the plot.
plt.colorbar()
plt.axis(aspect='image');
# plt.axis(aspect='image') to make x and y units match
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
# over-plot contours with labels on the contours themselves
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)

plt.colorbar();
```

#### Histogram,binnings,density

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

data= np.random.randn(1000)
plt.hist(data);
```

```python
plt.hist(data,bins=30,normed=True,alpha=0.8,
        histtype='stepfilled',color='blue',
        edgecolor='none');
```

```python
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)

kwargs= dict(histtype='stepfilled',alpha=0.3,normed=True,bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
counts,bin_edges = np.histogram(data,bins=5)
print(counts)
```

```python
#### Two dimensional Histogram and binnings,
mean=[0,0]
cov=[[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean,cov,size=10000).T

plt.hist2d(x,y,bins=30,cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

```python
from scipy.stats import gaussian_kde
# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)
# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))
# Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape),
            origin='lower', aspect='auto',
            extent=[-3.5, 3.5, -6, 6],
            cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
```
